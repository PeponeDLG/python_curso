```python
# Descargamos los datos de https://www.kaggle.com/datasets/rohitsahoo/sales-forecasting
# Los descomprimimos
# fichero 'SuperStore_Sales_Dataset.csv'
```

```python
import pandas as pd

df = pd.read_csv('SuperStore_Sales_Dataset.csv')
df
```

## Revisión de datos

```python
# Examinamos los datos
print("Primeras 5 filas:\n", df.head())
```

```python
print("Información general:\n")
print("Forma del dataset:", df.shape)  # Ej: (9994, 17)
print("Columnas:", df.columns.tolist())
print("\nTipos de datos:\n", df.dtypes)
```

```python
print("\nEstadísticas descriptivas de columnas numéricas:\n", df.describe())
```

## Limpieza de datos

```python
# Limpieza de datos
print("Duplicados:", df.duplicated().sum())
```

```python
print("Valores nulos por columna:\n", df.isnull().sum())
```

```python
# Convertir fechas
df['Order Date'] = pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df
```

## Detectar valores atípicos

```python
# Detectar outliers en Sales (usando IQR)
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 1.5 * IQR
outliers_sales = df[df['Sales'] > limite_superior]
print("Outliers en Sales:", len(outliers_sales))  # Ej: ~500
outliers_sales
```

```python
# Reemplazar outliers por el límite superior (o eliminar, según criterio)
df['Sales'] = df['Sales'].clip(upper=limite_superior)
```

```python
# Crear columna de Profit Margin (para análisis)
df['Profit Margin'] = df['Profit'] / df['Sales'] * 100  # En porcentaje
```

```python
# Resultado final
print("Dataset limpio - Forma:", df.shape)
print("Primeras filas después de limpieza:\n", df.head())
```

## Análisis exploratorio y estadístico

```python
# Estadísticas descriptivas por categoría
print("Ventas por Categoría:\n", df.groupby('Category')['Sales'].agg(['mean', 'sum', 'count']))
```

```python
# Correlación entre Sales y Profit
from scipy.stats import pearsonr, linregress, ttest_ind, chi2_contingency
corr, p_value = pearsonr(df['Sales'], df['Profit'])
print(f"Correlación Sales-Profit: {corr:.3f}, p-valor: {p_value:.3f}")
```

```python
# Regresión lineal Sales vs Profit
slope, intercept, r_value, p_value, std_err = linregress(df['Sales'], df['Profit'])
print(f"R²: {r_value**2:.3f}")
```

```python
# T-test: Comparar ganancias promedio por región (ej. Central vs West)
profit_central = df[df['Region'] == 'Central']['Profit']
profit_west = df[df['Region'] == 'West']['Profit']
t_stat, p_value_t = ttest_ind(profit_central, profit_west)
print(f"T-test regiones: t={t_stat:.3f}, p={p_value_t:.3f}")
```

```python
# Tabla de contingencia y Chi-cuadrado: Segment vs Category
tabla = pd.crosstab(df['Segment'], df['Category'])
chi2, p_value_chi, dof, expected = chi2_contingency(tabla)
print(f"Chi-cuadrado: {chi2:.3f}, p={p_value_chi:.3f}")
```

## Visualización de patrones

```python
# Gráfico de barras: Ventas por Región
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Region', y='Sales', estimator='mean')
plt.title('Ventas Promedio por Región')
plt.ylabel('Ventas Promedio ($)')
plt.show()
```

```python
# Boxplot: Profit por Categoría
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit')
plt.title('Distribución de Ganancias por Categoría')
plt.show()
```

```python
# Heatmap de correlaciones
corr_matrix = df[['Sales', 'Profit', 'Quantity']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```
