```python
# Dibujando funciones matemáticas
# nos aseguramos de tener instalada numpy
!pip install numpy

# dibujando funciones
import matplotlib.pyplot as plt
import numpy as np
# 50 puntos equiespaciados entre el mínimo y el máximo y lo guarda en un array de numpy
x = np.linspace(-2,2,50) 

y = x**2 # y será una "lista" con los valores de x al cuadrado

plt.plot(x, y)

# Le añadimos título, etiquetas y una rejilla
plt.title('Función cuadrado')
plt.xlabel('x')
plt.ylabel('y = x**2')
plt.grid(True)
plt.show()
```

```python
# Podemos añadir varios gráficos en los mismos ejes
z = y + 2
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
# Podemos seleccionar el aspecto de cada gráfica, añadiendo un tercer argumento en la llamada a plot
w = z - 5

plt.plot(x,y,'r')   # línea rojo 
plt.plot(x,z,'g--') # línea gris discontinua
plt.plot(x,w,'b.')  # azul punteada

plt.plot(x,y,'r^')
plt.plot(x,z,'g:')
plt.plot(x,w,'bx')
plt.show()
```

```python
# Personalización y subgráficos
# Crear una figura con subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # 1 fila, 2 columnas
data = {'Día': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'],
        'Ventas': [100, 120, 90, 150, 130]}
df = pd.DataFrame(data)
# Gráfico de líneas en el primer subplot
ax1.plot(df['Día'], df['Ventas'], marker='o', linestyle='--', color='b', label='Ventas')
ax1.set_title('Ventas Diarias')
ax1.set_xlabel('Día')
ax1.set_ylabel('Ventas ($)')
ax1.legend()
ax1.grid(True)

# Histograma en el segundo subplot
ax2.hist(ventas, bins=5, color='purple', edgecolor='black')
ax2.set_title('Distribución de Ventas')
ax2.set_xlabel('Ventas ($)')
ax2.set_ylabel('Frecuencia')

plt.tight_layout()  # Ajusta el espaciado
plt.show()
```

```python
# Gráficos en 3D
# Ejemplo de https://matplotlib.org/stable/gallery/mplot3d/lorenz_attractor.html
import matplotlib.pyplot as plt
import numpy as np


def lorenz(xyz, *, s=10, r=28, b=2.667):
    """
    Parameters
    ----------
    xyz : array-like, shape (3,)
       Point of interest in three-dimensional space.
    s, r, b : float
       Parameters defining the Lorenz attractor.

    Returns
    -------
    xyz_dot : array, shape (3,)
       Values of the Lorenz attractor's partial derivatives at *xyz*.
    """
    x, y, z = xyz
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return np.array([x_dot, y_dot, z_dot])


dt = 0.01
num_steps = 10000

xyzs = np.empty((num_steps + 1, 3))  # Need one more for the initial values
xyzs[0] = (0., 1., 1.05)  # Set initial values
# Step through "time", calculating the partial derivatives at the current point
# and using them to estimate the next point
for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt

# Plot
ax = plt.figure().add_subplot(projection='3d')

ax.plot(*xyzs.T, lw=0.5)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor")

plt.show()
```

```python
# PairPlot
import seaborn as sns

# Dataset simulado
data = {
    'Población': [3.3e6, 1.6e6, 0.7e6, 1.2e6],
    'Temperatura': [15, 18, 20, 17],
    'Ventas': [1000, 800, 600, 900]
}
df = pd.DataFrame(data, index=['Madrid', 'Barcelona', 'Sevilla', 'Valencia'])

# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# HeatMap
# Calcular matriz de correlación
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Mapa de Correlación')
plt.show()
```

```python
# Boxplot

# Boxplot de Ventas por Ciudad
data = {
    'Ciudad': ['Madrid', 'Madrid', 'Barcelona', 'Barcelona', 'Sevilla'],
    'Ventas': [1000, 1200, 800, 850, 600]
}
df_ventas = pd.DataFrame(data)

sns.boxplot(x='Ciudad', y='Ventas', data=df_ventas, palette='Set2')
plt.title('Ventas por Ciudad')
plt.show()
```

```python
# Visualizando DataFrames
# Gráfico de barras directo
df_ventas.groupby('Ciudad')['Ventas'].sum().plot(kind='bar', title='Ventas Totales por Ciudad', color='teal')
plt.ylabel('Ventas ($)')
plt.show()

# Histograma directo
df['Ventas'].plot(kind='hist', bins=5, title='Distribución de Ventas', color='orange')
plt.xlabel('Ventas ($)')
plt.show()
```
